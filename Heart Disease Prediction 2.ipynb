### Importing Dependencies

### Feature Selection

```python
sns.set_context("paper",font_scale=1.7)
#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(25,25))
#plot heat map

g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="Blues")
```

### Splitting the features and Target

```python
X = df.drop(columns='target',axis=1)
Y = df['target']
```

```python
X
```

```python
Y
```

### Splitting into training data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model Training


```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model

model.fit(X_train,Y_train)
```

### Model Evalution

```python
# Accuracy Score on Training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:- ',training_data_accuracy)
```

```python
# Accuracy Score on Testing data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:- ',testing_data_accuracy)
```

### Building a Predictive System

```python
input_data = (37,1,2,130,250,0,1,187,0,3.5,0,0,2)

# change the input data to numpy array
array = np.asarray(input_data)

# reshape the numpy array as we are predicting for only 1 instance
array_reshaped = array.reshape(1,-1)

prediction = model.predict(array_reshaped)
print(prediction)

if(prediction[0]==0):
    print('The person does not have heart disease')
else:
    print('The person have heart disease')
```
